#### Q. Plot count of genre2 on a line chart where count>1 

```python
q1 = df['genre_2'].value_counts()
q1 = q1[q1>1]
q1
```

```python
print(q1.index)
print(q1.values)
```

# PLOT CHART

```python
plt.figure(figsize = (9,4))
plt.plot(q1.index, q1.values, color = 'Black', marker = 'd',label = 'COUNT',
         markersize = 8,lw = 2, linestyle = '-.')
# Q1.index = Index(['Drama', 'Adventure', 'Comedy', 'Romance', 'Biography', 'Action','Thriller', 'Fantasy', 'Mystery', 'Sci-Fi']
# Q1.value = [35 22  7  5  5  4  3  3  2  2]
plt.xlabel('Genre 2', color = "green")
plt.ylabel('Count', color = "green")
plt.grid()
plt.title("Genre 2 vs Count", fontsize = '20', color = 'maroon')
plt.xticks(rotation = 0)
plt.legend()
plt.show()
```

## Q. Plot year wise mean of profit on a line chart where<br> 
<br> 

<b> $Profit  = Gross - Budget$</b>

#### Converting budget and gross from dollars to million dollars

```python
d1['Gross'] = d1['Gross']/1000000
d1['budget'] = d1['budget']/1000000
d1.head()
```

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head()
```

```python
q2 = d1.groupby('title_year')['Profit'].mean().reset_index()
q2 = q2.sort_values(by = 'title_year', ascending=False)
q2
```

```python
plt.plot(q2['title_year'],q2['Profit'],color = 'brown', marker = 'd',label = 'Profit')
plt.xticks(rotation = 10)
plt.xlabel('TITLE_YEAR',color = 'green')
plt.ylabel('PROFIT',color = 'green')
plt.title('YEAR WISE MEAN OF PROFIT', fontsize = 20,  color = 'black' )
plt.legend()

plt.grid()
plt.show
```

#### Q. Plot year wise mean of budget and profit on a line chart where<br> 
<br> 
<b> $Profit  = Gross - Budget$</b>

```python
q3 = d1.groupby('title_year')['budget','Profit'].mean().reset_index()
q3 = q3.sort_values(by = 'title_year', ascending=False)
q3
```

```python
plt.plot(q2['title_year'],q2['Profit'],color = 'violet', marker = 'd',label = 'Profit')
plt.plot(q3['title_year'],q3['budget'],color = 'maroon', marker = '.',label = 'budget')
plt.xticks(rotation = 20)
plt.xlabel('TITLE_YEAR',color = 'green')
plt.ylabel('PROFIT',color = 'green')
plt.title('YEAR WISE MEAN OF PROFIT', fontsize = 20,  color = 'black' )
plt.legend()
plt.grid()
plt.show()
```
