
# PIE CHART

#### Represent the percentage distrubution of the year wise mean of profit on pie chart.
where, <b>$Profit = Gross - Budget$</b>

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head()
```

```python
q3 = d1.groupby('title_year')['Profit'].mean().reset_index()
q3
```

```python
plt.pie(q3['Profit'],labels = q3['title_year'], autopct = '%.2f.%%', shadow = True)
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.pie(q3['Profit'],labels = q3['title_year'],
        colors = ['blue', 'red', 'lightgreen', 'grey', 'violet', 'maroon','pink'] ,autopct = '%.2f.%%', shadow = True,
        explode = (0,0,0,0,0,0,0.2), counterclock = False)
plt.title("THE YEAR WISE PROFIT OF MEAN", fontsize = 16, color = 'darkgreen')

plt.show()
```
