#  Bar plot :
<b>IQR</b> = Q3-Q1<br>
<b>MIN</b> = Q1- 1.5 IQR <br>
<b>MAX</b> = Q3 + 1.5 IQR<br>
<b>OUTLIERS</b> = >MAX OR <MIN
                        

#### Demo Example:

```python
x = [12,14,16,15,14,25,12,11,9,10,13,15,2,17,9,16]
print(x)
print(len(x))
```

```python
plt.boxplot(x)
plt.grid()
plt.show()
```

```python
q1 = np.percentile(x,25)
q2 = np.percentile(x,50) # mediun
q3 = np.percentile(x,75) 

iqr = q3 - q1
bmax = q3 + 1.5*iqr
bmin = q1 - 1.5*iqr

print('Q1', q1)
print('Q2', q2)
print('Q3', q3)
print('IQR', iqr)
print('bmax', bmax)
print('bmin', bmin)
```

###  q. plot the mean of budget and profit based on genre 1 on stack bar chart

```python
r1 =d1.groupby('genre_1')['budget','Profit'].mean()
r1.index.names = ['Genre']
```

```python
plt.bar(r1.index, d1['Profit'], color='red', label='Mean of Profit', bottom = r1['Profit'])
plt.bar(r1.index, d1['budget'], color='blue', label='Mean of Budget')
plt.xlabel('Genre_1', color='green')
plt.ylabel('Budget & Profit', color='green')
plt.legend()
plt.title('Genre Wise Mean of Budget & Profit')
plt.xticks(rotation=90)

plt.ylim(0,280)
plt.show()
```
